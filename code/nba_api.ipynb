```python
import numpy as np
import pandas as pd
from nba_api.stats.static import players
from nba_api.stats.endpoints import playercareerstats
import matplotlib.pyplot as plt
import seaborn as sns

# Steph Curry
career = playercareerstats.PlayerCareerStats(player_id='201939') 

pd.set_option('display.max_columns', None)  
career.get_data_frames()[0]
```

```python
from nba_api.stats.endpoints import shotchartdetail
def find_player(player_id):
    shots = shotchartdetail.ShotChartDetail(
        team_id=0,
        player_id=player_id ,  # Steph Curry
        season_type_all_star='Regular Season',
        context_measure_simple='FGA'
    ).get_data_frames()[0]
    return shots
shots = find_player(201939)
shots
```

Data for Steph Curry

```python
curry = shots.drop(columns=['GRID_TYPE', 'GAME_ID', 'GAME_EVENT_ID', 'PLAYER_ID', 'TEAM_ID', 'TEAM_NAME', 'PLAYER_NAME', 'EVENT_TYPE', 'GAME_DATE',	
                            'HTM',	'VTM'])
curry = curry.dropna()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=curry, x='LOC_X', y='LOC_Y', hue='SHOT_ZONE_BASIC', alpha=0.6, edgecolor=None)
plt.title('Steph Curry Shot Chart')
plt.xlim(-250, 250)
plt.ylim(-50, 420)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
```

```python

plt.figure(figsize=(12, 8))
sns.scatterplot(data=curry, x='LOC_X', y='LOC_Y', hue='SHOT_MADE_FLAG', palette = 'coolwarm', alpha=0.8, edgecolor=None)
plt.title('Steph Curry Shot Chart')
plt.xlim(-250, 250)
plt.ylim(-50, 420)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
```

```python
zone = curry.groupby('SHOT_ZONE_AREA').agg({'SHOT_MADE_FLAG': 'sum',
                                           'SHOT_ATTEMPTED_FLAG': 'count'}).sort_values(by='SHOT_MADE_FLAG').reset_index()
melted = zone.melt(id_vars='SHOT_ZONE_AREA', value_vars=['SHOT_MADE_FLAG', 'SHOT_ATTEMPTED_FLAG'], var_name='shot_type', value_name='count')
plt.figure(figsize=(12, 6)) 
sns.barplot(data=melted, x='SHOT_ZONE_AREA', y='count', hue = 'shot_type', palette='Set1')
plt.title('Steph Curry Shot made by Zone')
plt.xticks(rotation=45)
plt.xlabel('Shot Zone Area')
plt.ylabel('Shots')
plt.tight_layout()
plt.show()
```

```python
time = curry.sort_values(by=['MINUTES_REMAINING','SECONDS_REMAINING']).copy()
time['remaining'] = time['MINUTES_REMAINING'] * 60 + time['SECONDS_REMAINING']
plt.figure(figsize=(12, 6))
sns.scatterplot(data=time, x='remaining', y='SHOT_DISTANCE', hue='SHOT_MADE_FLAG', palette='coolwarm', alpha=0.7, edgecolor=None)
plt.title('Steph Curry time remaining vs Shot Distance')
plt.xlabel('Seconds Remaining')
plt.ylabel('Distance from Basket (feet)')
plt.legend()
plt.show()
```
